```python
##### (1) 

# Neighbor size를 정해서 예측치를 계산하는 함수 
import numpy as np
from scipy.stats import mode

def cf_knn_binary(user_id, place_id, neighbor_size=0):
    if place_id in likes_matrix:
        # 현재 사용자와 다른 사용자 간의 similarity 가져오기
        sim_scores = user_similarity[user_id].copy()
        
        # 현재 장소에 대한 모든 사용자의 heart 값 가져오기
        place_likes = likes_matrix[place_id].copy()
        
        # 현재 영화를 평가하지 않은 사용자의 index 가져오기
        none_likes_idx = place_likes[place_likes.isna()].index
       
        # 현재 영화를 평가하지 않은 사용자의 likes (null) 제거
        place_likes = place_likes.drop(none_likes_idx)
        
        # 현재 영화를 평가하지 않은 사용자의 similarity값 제거
        sim_scores = sim_scores.drop(none_likes_idx)
        
        if neighbor_size == 0:          
            # 이웃의 가장 일반적인 평점(mode)을 사용하여 예측
            predicted_rating = mode(place_likes).mode[0]
        else:                       
            if len(sim_scores) > 1: 
                # 지정된 이웃 크기 값과 해당 장소를 평가한 총 사용자 수 중 작은 것으로 결정
                neighbor_size = min(neighbor_size, len(sim_scores))
                # 배열로 변환 (argsort를 사용하기 위함)
                sim_scores = np.array(sim_scores)
                place_likes = np.array(place_likes)
                
                # 유사도를 순서대로 정렬
                user_idx = np.argsort(sim_scores)
                
                # 유사도를 이웃 크기만큼 받기
                sim_scores = sim_scores[user_idx][-neighbor_size:]
                
                # 영화 rating을 이웃 크기만큼 받기
                place_likes = place_likes[user_idx][-neighbor_size:]
                
                # 이웃의 가장 일반적인 평점(mode)을 사용하여 예측
                predicted_rating = mode(place_likes).mode[0]
            else:
                # 최소한 2명의 이웃이 필요한 경우에만 예측을 수행할 수 있도록
                predicted_rating = 0
    else:
        # 만약 장소가 평가되지 않았다면 기본적으로 0으로 예측
        predicted_rating = 0
    
    return predicted_rating


# 정확도 계산
score(cf_knn_binary, neighbor_size=0)
```

```python
def ubcf_binary(user_similarity, user_id):
    
    # 현재 사용자와 다른 사용자 간의 유사도 가져오기
    similar_users = user_similarity[user_id].copy()
    #print(similar_users)
    
    #
    similar_users = similar_users[similar_users > 0].index
    print("현재 사용자와 유사한 사용자", similar_users)
    
    # 사용자의 선호도 정보 가져오기
    user_likes = (matrix_dummy.loc[user_id] == 1).astype(int)
    #print("user_likes")
    #print(user_likes)
    # 현재 사용자와 유사한 사용자들의 찜한 장소 중에서 현재 사용자가 찜하지 않은 장소 찾기
    unliked_places = user_likes[user_likes == 0].index
    #print("unliked_places", unliked_places)
    
    # 유사한 사용자들의 찜한 장소를 기반으로 현재 사용자의 찜할 장소 예측
    predicted_likes = user_similarity.loc[similar_users, user_id].dot(matrix_dummy.loc[similar_users, unliked_places])
    predicted_likes = predicted_likes / user_similarity.loc[similar_users, user_id].sum()
    return predicted_likes

# 예측 찜 구하기
user_id = 3
predicted_likes = ubcf_binary(user_similarity, user_id)

print(f"사용자 {user_id}이 찜할 장소 예상:")
predicted_likes[predicted_likes>0].sort_values(ascending=False).head(5)
```

```python
import numpy as np

def ubcf_binary_recommendation(user_similarity, user_id, threshold=0.5):
    # 현재 사용자와 다른 사용자 간의 similarity 가져오기
    sim_scores = user_similarity[user_id].copy()
    
    # 현재 사용자가 이미 평가한 영화 가져오기 (이진 데이터에서는 1인 영화만 고려)
    user_ratings = (matrix_dummy.loc[user_id] > 0).astype(int)
    
    # 현재 사용자가 이미 평가한 영화는 추천에서 제외하기 위해 유사도에서 해당 영화의 값을 0으로 설정
    sim_scores[user_ratings.index] = 0
    
    # 유사도가 threshold 이상인 사용자 중에서 1로 평가한 영화의 비율을 계산
    weighted_ratings = (user_similarity[user_id] > threshold) * user_ratings
    scores = weighted_ratings.sum() / (user_similarity[user_id] > threshold).sum()
    
    return scores

# 특정 사용자에게 영화를 추천
user_id = 1
recommendation = ubcf_binary_recommendation(user_similarity, user_id)

print(f"사용자 {user_id}에게 추천할 영화의 예상 평가: {recommendation}")
```

```python


# Neighbor size를 정해서 예측치를 계산하는 함수
def cf_binary_knn(user_id, place_id, neighbor_size=0, threshold=0.5):
    if place_id in rating_matrix:
        # 현재 사용자와 다른 사용자 간의 similarity 가져오기
        sim_scores = user_similarity[user_id].copy()
        # 현재 장소에 대한 모든 사용자의 하트 정보 가져오기
        place_ratings = rating_matrix[place_id].copy()
        # 현재 장소를 평가하지 않은 사용자의 index 가져오기
        none_rating_idx = place_ratings[place_ratings == 0].index
        # 현재 장소를 평가하지 않은 사용자의 하트 정보 제거
        place_ratings = place_ratings.drop(none_rating_idx)
        # 현재 장소를 평가하지 않은 사용자의 similarity값 제거
        sim_scores = sim_scores.drop(none_rating_idx)

        # 이진 데이터에서는 하트 정보가 threshold 이상인 경우 1, 그렇지 않으면 0으로 처리
        place_ratings = (place_ratings >= threshold).astype(int)

        if neighbor_size == 0:
            # 현재 장소를 평가한 모든 사용자의 가중평균값 구하기
            mean_heart = np.dot(sim_scores, place_ratings) / sim_scores.sum()
        else:
            if len(sim_scores) > 1:
                neighbor_size = min(neighbor_size, len(sim_scores))
                sim_scores = np.array(sim_scores)
                place_ratings = np.array(place_ratings)
                user_idx = np.argsort(sim_scores)
                sim_scores = sim_scores[user_idx][-neighbor_size:]
                place_ratings = place_ratings[user_idx][-neighbor_size:]
                mean_heart = np.dot(sim_scores, place_ratings) / sim_scores.sum()
            else:
                mean_heart = 0.5  # 사용자가 유사한 다른 사용자가 없는 경우 0.5로 설정 (임의의 기본값)
    else:
        mean_heart = 0.5
    return mean_heart
```
