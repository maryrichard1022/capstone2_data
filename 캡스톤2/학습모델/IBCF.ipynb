```python
import pandas as pd
import numpy as np
from scipy.stats import mode
from sklearn.metrics import f1_score
from sklearn.metrics import jaccard_score
from sklearn.model_selection import train_test_split
from sklearn.metrics.pairwise import cosine_similarity

# ê° í–‰ ê°„ì˜ Jaccard ìœ ì‚¬ë„ ê³„ì‚°
def calculate_jaccard_similarity(matrix):
    num_users = matrix.shape[0]
    similarities = np.zeros((num_users, num_users))

    for i in range(num_users):
        for j in range(i, num_users):
            similarity = jaccard_score(matrix.iloc[i], matrix.iloc[j])
            similarities[i, j] = similarity
            similarities[j, i] = similarity

    return similarities



likes = pd.read_csv(f"C:/Users/ê¹€ê°€ì—°/Desktop/23-2í•™ê¸°/ìº¡ìŠ¤í†¤2/data/hearts_data.csv", encoding='UTF-8-SIG')
likes = likes[["place_id", "user_id"]]
likes.loc[:, "heart"] = 1
likes = likes.drop_duplicates(subset=['place_id', 'user_id'], keep='first')


x = likes.copy()
y = likes['user_id']


x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, stratify=y)

#  train ë°ì´í„°ë¡œ Full matrix êµ¬í•˜ê¸° 
likes_matrix = x_train.pivot(index='user_id', columns='place_id', values='heart')
likes_matrix
```

```python
likes_matrix_t = likes_matrix.transpose()
likes_matrix_t
```

```python
# train set ì‚¬ìš©ìžë“¤ì˜ Cosine similarities ê³„ì‚°
matrix_dummy_t = likes_matrix_t.copy().fillna(0)
```

```python
# ì½”ì‚¬ì¸ ìœ ì‚¬ë„
# train set ì‚¬ìš©ìžë“¤ì˜ Cosine similarities ê³„ì‚°
item_similarity = cosine_similarity(matrix_dummy_t, matrix_dummy_t)
item_similarity = pd.DataFrame(item_similarity, index=likes_matrix_t.index, columns=likes_matrix_t.index)
item_similarity
```

### Item-Based Collaborative Filtering ì•Œê³ ë¦¬ì¦˜
#### ì£¼ì–´ì§„ ì‚¬ìš©ìžì™€ ìž¥ì†Œì— ëŒ€í•´ í•´ë‹¹ ìž¥ì†Œì˜ ì˜ˆì¸¡ì¹˜ë¥¼ ê³„ì‚°

```python
# ì •í™•ë„ë¥¼ ê³„ì‚°í•˜ëŠ” í•¨ìˆ˜ 
def f1(y_true, y_pred):
    return f1_score(y_true, y_pred)

# ëª¨ë¸ë³„ ì •í™•ë„ë¥¼ ê³„ì‚°í•˜ëŠ” í•¨ìˆ˜ 
def score(model):
    id_pairs = zip(x_test['user_id'], x_test['place_id'])
    
    # ì˜ˆì¸¡ ê°’
    y_pred = np.array([model(user, place) for (user, place) in id_pairs]).astype(int)
    #print("y_pred", y_pred)
    
    # ì‹¤ì œ ê°’
    y_true = np.array(x_test['heart'])
    #print("y_true", y_true)
    return f1(y_true, y_pred)
```

```python
def CF_IBCF(user_id, place_id):
    if place_id in item_similarity:      # í˜„ìž¬ ì˜í™”ê°€ train setì— ìžˆëŠ”ì§€ í™•ì¸
        # í˜„ìž¬ ìž¥ì†Œì™€ ë‹¤ë¥¸ ìž¥ì†Œì˜ similarity ê°’ ê°€ì ¸ì˜¤ê¸°
        sim_scores = item_similarity[place_id]
        #print(sim_scores)
        # í˜„ ì‚¬ìš©ìžì˜ ëª¨ë“  ì°œ ê°’ ê°€ì ¸ì˜¤ê¸°
        user_likes = likes_matrix_t[user_id]
        #print(user_likes)
        
        
        # ì‚¬ìš©ìžê°€ í‰ê°€í•˜ì§€ ì•Šì€ ìž¥ì†Œ index ê°€ì ¸ì˜¤ê¸°
        none_likes_idx = user_likes[user_likes.isnull()].index
        #print(none_likes_idx)
        # ì‚¬ìš©ìžê°€ í‰ê°€í•˜ì§€ ì•Šì€ ìž¥ì†Œ ì œê±°
        user_likes = user_likes.dropna()
        #print(user_likes)
        # ì‚¬ìš©ìžê°€ í‰ê°€í•˜ì§€ ì•Šì€ ìž¥ì†Œì˜ similarity ê°’ ì œê±°
        sim_scores = sim_scores.drop(none_likes_idx)
        #print(sim_scores)
        
        # í˜„ ìž¥ì†Œì— ëŒ€í•œ ì˜ˆìƒ ì°œ ê³„ì‚°, ê°€ì¤‘ì¹˜ëŠ” í˜„ ìž¥ì†Œì™€ ì‚¬ìš©ìžê°€ í‰ê°€í•œ ìž¥ì†Œì˜ ìœ ì‚¬ë„
        if sim_scores.sum() != 0.0:
            predicted_likes = np.dot(sim_scores, user_likes) / sim_scores.sum()
            #print(predicted_likes)
        else:
            predicted_likes = mode(user_likes, keepdims=True).mode[0]
            #print(predicted_likes)
    else:
        predicted_likes = 0.0
    return predicted_likes

# ì •í™•ë„ ê³„ì‚°
score(CF_IBCF)
```

```python
def IBCF_recommender(user, n_items):
    # í˜„ìž¬ ì‚¬ìš©ìžê°€ ì°œí•œ ìž¥ì†Œ
    liked_index = likes_matrix.loc[user][likes_matrix.loc[user] > 0].index
    
    # ëª¨ë“  ìž¥ì†Œì— ëŒ€í•œ ì˜ˆìƒ ì°œ ì—¬ë¶€ ê³„ì‚°
    predictions = []
    for place_id in likes_matrix.columns:
        if place_id not in liked_index:
            prediction = CF_IBCF(user, place_id)
            #print("ì˜ˆìƒ", prediction)
            predictions.append((place_id, prediction))
    
    # ì˜ˆìƒ ì°œ ì—¬ë¶€ë¥¼ ê¸°ì¤€ìœ¼ë¡œ ë‚´ë¦¼ì°¨ìˆœ ì •ë ¬
    predictions.sort(key=lambda x: x[1], reverse=True)
    print("predictions", predictions)
    # ìƒìœ„ n_itemsê°œì˜ ìž¥ì†Œë¥¼ ì¶”ì²œ
    recommended_items = [place_id for place_id, _ in predictions[:n_items]]
    #print("ì¶”ì²œ ìž¥ì†Œ", recommended_items)
    return recommended_items

# ì‚¬ìš©ìžì—ê²Œ IBCF ì•Œê³ ë¦¬ì¦˜ì„ ì‚¬ìš©í•˜ì—¬ ìµœëŒ€ 10ê°œì˜ ìž¥ì†Œë¥¼ ì¶”ì²œ
IBCF_recommender(user=52, n_items=100)
```

```python
def recommender(user, n_items):
    # í˜„ìž¬ ì‚¬ìš©ìžì˜ ëª¨ë“  ì•„ì´í…œì— ëŒ€í•œ ì˜ˆìƒ í‰ì  ê³„ì‚°
    predictions = [] # ì˜ˆìƒ í‰ì ì„ ì €ìž¥í•  ë¦¬ìŠ¤íŠ¸
    liked_index = likes_matrix[user][likes_matrix[user] > 0].index  # ì´ë¯¸ ì°œí•˜ê¸° ëˆ„ë¥¸ ìž¥ì†Œ ì œì™¸í•˜ê³  ì¶”ì²œí•˜ê¸° ìœ„í•´ ê°€ì ¸ì˜´
    #print("liked_index", liked_index)
    items = likes_matrix[user].drop(liked_index) # ì°œí•˜ê¸° ëˆ„ë¥¸ ìž¥ì†Œ ì œì™¸
    #print(items)
    for item in items.index: # ê° ì•„ì´í…œì— ëŒ€í•œ í˜„ ì‚¬ìš©ìžì˜ ì˜ˆìƒ í‰ì  ê³„ì‚°
        predictions.append(CF_IBCF(user, item)) # í˜„ìž¬ ì•„ì´í…œì— ëŒ€í•œ ì˜ˆìƒ í‰ì  êµ¬í•˜ê¸°
    recommendations = pd.Series(data=predictions, index=items.index, dtype=float) # ì˜ˆìƒ í‰ì ì„ ì‹œë¦¬ì¦ˆë¡œ ë³€í™˜
    #print(recommendations)
    #recommendations = recommendations.sort_values(ascending=False)[:n_items] # ì˜ˆìƒ í‰ì  ë†’ì€ ìˆœì„œë¡œ ì •ë ¬í•´ì„œ nê°œë§Œí¼ ë½‘ìŒ
    recommendations = recommendations[recommendations == 1]
    return recommendations

recommender(user=7, n_items=100)
```
