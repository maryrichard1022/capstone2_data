하버사인 공식을 사용하여 두 지점간의 거리 계산하기
-> 지구 표면에서 두 지점 간의 직선 거리를 계산하는데 사용

```python
from math import sin, cos, sqrt, atan2

def haversine_distance(x0, y0, x1, y1):
    # x, y가 빈 값이 아닌 경우에만 함수 실행
    if x0 is not None and y0 is not None:
        R = 6371  # 지구의 반지름. 킬로미터 단위.
        dlat = x1 - x0
        dlon = y1 - y0

        # 하버사인 공식 사용
        a = sin(dlat / 2)**2 + cos(x0) * cos(y0) * sin(dlon / 2)**2
        c = 2 * atan2(sqrt(a), sqrt(1 - a))

        distance = R * c

        return distance
    else:
        return None
```

```python
# 기존의 places DataFrame을 복사하여 새로운 DF 생성
distance = places[["id", "name"]].copy()

# 거리 계산하여 생성한 DF에 값을 저장함
distance["distances"] = places.apply(lambda row: haversine_distance(x, y, row["latitude"], row["longitude"]), axis=1)
distance
```
