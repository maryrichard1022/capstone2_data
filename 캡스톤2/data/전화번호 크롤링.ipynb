```python
import requests
from bs4 import BeautifulSoup
import re

# 장소명 리스트

# 전화번호 패턴
phone_number_pattern = r'(02-\d{3,4}-\d{4}|010-\d{4}-\d{4})'

for selected_place_name in place_name:
    url = f"https://search.daum.net/search?nil_suggest=btn&w=tot&DA=SBC&q={selected_place_name}"

    res = requests.get(url)
    res.raise_for_status()
    html = res.text
    soup = BeautifulSoup(html, "html.parser")

    try:
        # 웹 페이지에서 전화번호 형식의 텍스트 추출
        phone_number_text = re.search(phone_number_pattern, html)

        if phone_number_text:
            phone_number = phone_number_text.group()
            # 추출한 번호를 출력하거나 데이터프레임에 저장
            print(f"{selected_place_name} 전화번호: {phone_number}")
            df.loc[df['검색어'] == selected_place_name, '전화번호'] = phone_number

    except Exception as e:
        print(f"{selected_place_name} 정보를 가져오는 중 오류 발생: {str(e)}")
```
