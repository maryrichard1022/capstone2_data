```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("문화공간 new.csv", encoding='CP949')
df.head()
```

```python
import re # 정규 표현식 사용

 # 정규식을 사용하여 "구"로 끝나는 단어 추출하여 "자치구" 열에 저장
df['district'] = df['address'].str.extract(r'(\w*구\w*)')
df['district']
```

```python
df.head()
```

```python
# DataFrame을 CSV 파일로 저장 (인덱스 포함, utf-8 인코딩)
df.to_csv('문화공간 new 자치구.csv', index=True, encoding='CP949')
```

```python
df["phone_number"]
```

```python
# 전화번호 ~부터 뒤에글자 없애기
df["phone_number"] = df["phone_number"].str.split('~').str.get(0)
df["phone_number"]
```

```python
# )는 -로 바꾸기
df['phone_number'] = df['phone_number'].str.replace(')', '-')
df["phone_number"]
```

```python
# "("를 제거하기
df['phone_number'] = df['phone_number'].str.replace('(', '')
df["phone_number"]
```

```python
"공백 없애기"
df['phone_number'] = df['phone_number'].str.replace(' ', '')
df["phone_number"]
```

```python

# 정보없음, 없음, 홈페이지 참고는 빈값으로 바꾸기
df['phone_number'].replace(['정보없음', '없음', '홈페이지 참고'], np.nan, inplace=True)
df['phone_number'].unique()
```

```python
df['phone_number'].str.contains('정보').any()
```

```python
# DataFrame을 CSV 파일로 저장 (인덱스 포함, utf-8 인코딩)
df.to_csv('문화공간 new 자치구.csv', index=True, encoding='CP949')
```
