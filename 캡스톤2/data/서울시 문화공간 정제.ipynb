```python
# 인코딩 방식 확인
import chardet
with open("서울시 생활체육포털 우리동네 프로그램.csv", 'rb') as f:
    result = chardet.detect(f.read())

print("Detected encoding:", result['encoding'])
```

```python
df = pd.read_csv("서울시 문화공간 정보.csv", encoding='CP949')
df
```

```python
df['자치구'] = ""
df
```

```python
import re # 정규 표현식 사용

# 정규식을 사용하여 "구"로 끝나는 단어 추출하여 "자치구" 열에 저장
df['자치구'] = df['주소'].str.extract(r'(\w*구\w*)')

# 결과 출력
print(df['자치구'])
df
```

```python
# DataFrame을 CSV 파일로 저장 (인덱스 포함, utf-8 인코딩)
df.to_csv('서울시 문화공간 정보 정제.csv', index=True, encoding='CP949')
```
