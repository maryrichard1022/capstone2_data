```python
import pandas as pd
import numpy as np
df = pd.read_csv("문화공간 work time 정제.csv", encoding='CP949')
df.head()
```

```python
df['description']
```

```python
df['description'] = df['description'].str.replace(r'<.*?>', '', regex=True)
df['description']
```

```python
df['description'] = df['description'].str.split('.').str[0]
df['description']
```

```python
# DataFrame을 CSV 파일로 저장 (인덱스 포함, utf-8 인코딩)
df.to_csv('문화공간 description 정제.csv', index=True, encoding='CP949')
```
