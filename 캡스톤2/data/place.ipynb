```python
import pandas as pd
import numpy as np
gallery = pd.read_csv("완료/도시갤러리 최종.csv", encoding='CP949')
```

```python
workout = pd.read_csv("완료/생활체육 최종.csv", encoding='CP949')
```

```python
bookstore = pd.read_csv("완료/책방 최종.csv", encoding='CP949')
```

```python
event = pd.read_csv("완료/문화행사 최종.csv", encoding='CP949')
```

```python
space = pd.read_csv("완료/문화공간 최종.csv", encoding='CP949')
```

```python
future = pd.read_csv("완료/미래유산 최종.csv", encoding='CP949')
```

```python
total_place = pd.concat([gallery, workout, bookstore, event, space, future], join="inner", ignore_index=True)
total_place
```

```python
values_to_exclude = ['청소년', '시니어', '노인', '아동']
total_place[total_place['name'].str.contains('|'.join(values_to_exclude), case=False, regex=True)].shape
```

```python
# '분류' 열의 값이 values_to_exclude를 포함하지 않는 행만 선택
total_place = total_place[~total_place['name'].str.contains('|'.join(values_to_exclude), case=False, regex=True)]
total_place
```

```python
# DataFrame을 CSV 파일로 저장 (인덱스 포함, utf-8 인코딩)
total_place.to_csv('place.csv', encoding='CP949')
```

```python
total_place[total_place["price"].isna()]
```

```python
# 결측값 수작업으로 넣기

for index, row in total_place[total_place["price"].isna()].iterrows():
    # '분류' 열의 값이 '축제'를 포함하는 경우
    user_input = input(f'원하는 price 값을 입력하세요 (현재 행: {row["name"]}): ')
    total_place.at[index, 'price'] = user_input  # 사용자가 입력한 값을 price로 설정
```
