### filter_id 분리하기

```python
# 새로운 DataFrame을 만듭니다.
new_df = pd.DataFrame(columns=['place_id', 'name', 'filter_id'])
new_df
```

```python
# 기존 DataFrame을 순회하면서 filter_id를 분리합니다.
for index, row in df.iterrows():
    place_id = row['place_id']
    name = row['name']

    # 쉼표로 구분된 문자열을 나누고 각 부분을 정수로 변환하여 리스트로 만듭니다.
    try:
        filter_ids = [int(fid) for fid in row['filter_id'].split(',')]
    except ValueError:
        print(f"Error: Unable to convert 'filter_id' to integers for row {index}. Check the format of the 'filter_id'.")
        continue

    # 새로운 DataFrame에 데이터를 추가합니다.
    for filter_id in filter_ids:
        new_df = new_df.append({'place_id': place_id, 'name': name, 'filter_id': filter_id}, ignore_index=True)

# 결과를 출력합니다.
print(new_df)
```

```python
new_df = new_df.drop_duplicates()
new_df
```
