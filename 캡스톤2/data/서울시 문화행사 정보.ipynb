```python
import chardet

with open("서울시 문화행사 정보.csv", 'rb') as f:
    result = chardet.detect(f.read())

print("Detected encoding:", result['encoding'])
```

```python
df = pd.read_csv("서울시 문화행사 정보.csv", encoding='CP949')
df.head()
```

```python
df['날짜/시간']
```

```python
from datetime import datetime

# '~' 이후의 날짜를 추출하여 비교
df['끝_날짜'] = df['날짜/시간'].str.split('~').str[1]
df['끝_날짜'] = pd.to_datetime(df['끝_날짜'])

# 현재 날짜를 가져오기
today = datetime.now()

df = df[df['끝_날짜'] > today]

# 필요한 열만 선택
df = df.drop(['끝_날짜'], axis=1)
print(df)
```

```python
# DataFrame을 CSV 파일로 저장 (인덱스 포함, utf-8 인코딩)
df.to_csv("서울시 문화행사 정보123.csv", index=True, encoding='CP949')
```
