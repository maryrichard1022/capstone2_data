```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("문화공간 카테고리 분류 전.csv", encoding='CP949')
df.head()
```

```python
df['분류'].unique()
```

```python
df.loc[df['분류'].str.contains('예술회관|공연|콘서트|뮤지컬|영화|연극|무용|클래식|국악|둑주|극장'), 'category_id'] = 2
df.loc[df['분류'].str.contains('전시|미술관|박물관|갤러리'), 'category_id'] = 3
df.loc[df['분류'].str.contains('체험|문화원|종교|교육|절'), 'category_id'] = 5
df.loc[df['분류'].str.contains('공원|운동|체육|광장'), 'category_id'] = 1
df.loc[df['분류'].str.contains('거리|구청|학교|명소|쇼핑|마을|시장|장소'), 'category_id'] = 7
df.loc[df['분류'].str.contains('도서관|서점'), 'category_id'] = 4
df.loc[df['분류'].str.contains('카페|식당|맛집'), 'category_id'] = 6
```

```python
df[df['category_id'].isna()]
```

```python
# 결측값 수작업으로 넣기

for index, row in df[df['category_id'].isna()].iterrows():
    # '분류' 열의 값이 '축제'를 포함하는 경우
    user_input = input(f'원하는 category_id 값을 입력하세요 (현재 행: {row["name"]}): ')
    df.at[index, 'category_id'] = user_input  # 사용자가 입력한 값을 category_id로 설정
```

```python
df[df['category_id'].isna()]
```

```python
df.drop('분류', axis=1, inplace=True)
```

```python
# DataFrame을 CSV 파일로 저장 (인덱스 포함, utf-8 인코딩)
df.to_csv('문화공간 카테고리 분류 완료.csv', index=True, encoding='CP949')
```
