```python
df[df['category_id'].isna()]
```

```python
# 결측값 수작업으로 넣기

for index, row in df[df['category_id'].isna()].iterrows():
    # '분류' 열의 값이 '축제'를 포함하는 경우
    user_input = input(f'원하는 category_id 값을 입력하세요 (현재 행: {row["name"]}): ')
    df.at[index, 'category_id'] = user_input  # 사용자가 입력한 값을 category_id로 설정
```

```python
# DataFrame을 CSV 파일로 저장 (인덱스 포함, utf-8 인코딩)
df.to_csv('문화공간 카테고리 다시.csv', index=True, encoding='CP949')
```
