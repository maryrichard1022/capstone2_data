```python
import pandas as pd
import numpy as np

df = pd.read_csv("완료/미래유산 이미지, 가격, 카테고리 있는 테이블.csv", encoding='CP949')
df2 = pd.read_csv("완료/미래유산 진행중.csv", encoding='CP949')
```

```python
merged_df = df.merge(df2, on='name', how='left')
merged_df
```

```python
merged_df['image_url'] = merged_df['image_url_y'].fillna(merged_df['image_url_x'])
merged_df = merged_df.drop(['image_url_x', 'image_url_y'], axis=1)


merged_df.to_csv("미래유산 이미지 확인.csv", index=True, encoding='CP949')
```
