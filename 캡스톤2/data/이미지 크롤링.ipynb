```python
import pandas as pd
import numpy as np
df = pd.read_csv("서울시 서울미래유산 체험 코스 정제.csv", encoding='CP949')
df.head()
```

```python
place_name = df["name"]
place_name.tolist()
place_name
```

```python
from selenium.common.exceptions import NoSuchElementException
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys

browser_path = 'C:/Users/김가연/Desktop/23-2학기/캡스톤2/chromedriver.exe'

# 웹 드라이버 생성
browser = webdriver.Chrome(browser_path)
browser.get("https://www.google.co.kr/imghp?hl=ko&authuser=0&ogbl")

for selected_place_name in place_name:
    elem = browser.find_element_by_name("q")
    elem.clear()
    elem.send_keys("서울미래유산", selected_place_name)
    elem.send_keys(Keys.RETURN)
    time.sleep(3)
    try:
        browser.find_elements_by_css_selector(".rg_i.Q4LuWd")[0].click()
        time.sleep(3)
        image_url = browser.find_element_by_css_selector(".sFlh5c.pT0Scc.iPVvYb").get_attribute("src")
        print(selected_place_name, "첫번째 시도 : ", image_url)
        response = input("이미지를 사용하시겠습니까? (Y/N): ")
        if response.upper() == 'N':
            raise NoSuchElementException("이미지 요소를 찾을 수 없음")
        elif response.upper() == 'Y':
            df.loc[df['name'] == selected_place_name, 'image_url'] = image_url
    except NoSuchElementException:
        try:
            browser.find_elements_by_css_selector(".rg_i.Q4LuWd")[1].click()
            time.sleep(3)
            image_url = browser.find_element_by_css_selector(".sFlh5c.pT0Scc.iPVvYb").get_attribute("src")
            print(selected_place_name, "두번째 시도 : ", image_url)
            response = input("이미지를 사용하시겠습니까? (Y/N): ")
            if response.upper() == 'N':
                raise NoSuchElementException("이미지 요소를 찾을 수 없음")
            elif response.upper() == 'Y':
                df.loc[df['name'] == selected_place_name, 'image_url'] = image_url
        except NoSuchElementException:
            print(selected_place_name, " 이미지 요소를 찾을 수 없음")
```

```python
# DataFrame을 CSV 파일로 저장 (인덱스 포함, utf-8 인코딩)
df.to_csv('서울 도시갤러리 정제 업데이트.csv', index=True, encoding='CP949')
```
